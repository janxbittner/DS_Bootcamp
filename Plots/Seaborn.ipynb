Import module:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Tips dataset:

Import data:

```python
df = sns.load_dataset('tips')
df.head(3)
```

Statistics about data:

```python
df.describe(include = 'category').T
```

```python
df.describe().T
```

```python
df.info()
```

Plot with tip and total bill by gender and size:

```python
_= sns.relplot(data= df,
           x = 'total_bill',
           y = 'tip',
           palette = 'rocket',
           hue = 'sex',
           size = 'size',
           legend = 'auto')
```

Tips and total bill by day of week:

```python
_= sns.relplot(data= df,
           x = 'total_bill',
           y = 'tip',
           palette = 'rocket',
           hue = 'sex',
           size = 'size',
           legend = 'auto',
           col = 'day',
           col_wrap = 2)
```

```python
_= sns.relplot(data= df,
           x = 'total_bill',
           y = 'tip',
           palette = 'rocket',
           hue = 'sex',
           size = 'size',
           legend = 'auto',
           col = 'day',
           row = 'time')
```

Total bill by day:

```python
_ = sns.catplot(data = df,
           x = 'day',
           y = 'total_bill',
           col = 'time',
           kind = 'violin')
```

# Titanic dataset

Load data:

```python
tit = sns.load_dataset('titanic')
tit.head(3)
```

Statistics of dataset:

```python
tit.info()
```

```python
tit.describe().T
```

```python
tit.describe(include ='category').T
```

```python
tit.describe(include ='object').T
```

Passengers by deck:

```python
_ = sns.catplot(data = tit,
               x = 'deck',
               kind = 'count',
               palette = 'rocket')
```

Chance to survive by deck:

```python
surv = tit.groupby(by = ['deck']).aggregate('mean')
surv = surv.survived.to_frame()
surv
```

```python
_ = sns.catplot(data = surv,
               x = surv.index,
               y = 'survived',
               palette = 'rocket',
               kind = 'bar')
```
