## Relational plots with relplot

Import modules

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style = 'whitegrid', palette = 'colorblind', font_scale= 1.1)
```

Load FMRI data set using seaborn module and show last 5 rows

```
df = sns.load_dataset('tips')
df.tail(n = 3)
```

Show information about columns type and values in it

```
df.info()
```

Show description of numerical data

```
df.describe()
```

Show description of text data

```
df.describe(include= ['category'])
```

Create category plot with days and total bills.

```
sns.catplot(data = df, x = 'day', y= 'total_bill')
```

Create category plot with statistics about days and total bills.

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'box')
```

Create category plot with statistics about days and total bills (type violin).

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'violin')
```

Create category plot with statistics about days and total bills (type boxen).

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'boxen')
```

Create category plot with statistics about days and total bills (type point).

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'point')
```

Create plot with confidence interval of total_bill and day.

```
sns.catplot(data = df, x = 'day', y = 'total_bill', kind = 'bar')
```

Create plot with numbers by customers by day using catplot.

```
sns.catplot(data = df, x = 'day', kind = 'count')
```

Create data frame without *Thursday*

```
end_days = df.drop(df[df.day == "Thur"].index)
end_days.day = end_days.day.cat.remove_categories(['Thur'])
```

Check if Thursday is removed correctly

```
end_days.day.value_counts()
```

Create plot which show total bills in Friday, Saturday and Sunday.

```
sns.catplot(data = end_days, x = 'day',y= 'total_bill',
            kind = 'swarm',hue = 'sex')
```
