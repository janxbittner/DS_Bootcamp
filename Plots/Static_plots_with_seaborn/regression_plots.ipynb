Importing modules.

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style = 'darkgrid',palette= 'viridis')
```

Import data set about restaurant tips and bills. Show first 3 rows.

```
df = sns.load_dataset('tips')
df.head(3)
```

Create regression plot about total bill and tip.

```
sns.regplot('total_bill','tip',data = df, x_estimator = np.mean)
```
