Importing modules.

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style = 'darkgrid',palette= 'copper_r')
```

Create table with multivariante normal distribution, then convert into data frame and show first 3 rows.

```
df = np.random.multivariate_normal(mean = [0,1], cov = [(1,-0.04),(-0.4,1)],
                                   size = 200)
df = pd.DataFrame(df,columns=['x','y'])
df.head(3)
```

Show statistics of dataframe.

```
df.describe()
```

Create plot of x,y data.



```
sns.jointplot(data = df, x = 'x',y = 'y')
```

Create plot of x,y data, kind of hex.

```
sns.jointplot(data = df,x ='x',y = 'y', kind = 'hex')
```

Create x,y plot with kernal density estimation.

```
sns.jointplot(data = df,x = 'x',y = 'y', kind = 'kde', n_levels =10)
```

Create joint plots with scatter and kernal distributed estimation.

```
kde_plot = sns.jointplot(data = df, x ='x', y='y', kind = 'kde')
kde_plot.plot_joint(plt.scatter, c = 'w', s =20)
```
