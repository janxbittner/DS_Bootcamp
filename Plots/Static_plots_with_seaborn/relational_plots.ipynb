## Relational plots with relplot

Import modules

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style = 'whitegrid', palette = 'colorblind', font_scale= 1.1)
```

Load FMRI data set using seaborn module and show last 5 rows

```
df = sns.load_dataset('fmri')
df.tail(n = 5)
```

Show information about columns type and values in it

```
df.info()
```

Show description of numerical data

```
df.describe()
```

Show description of text data

```
df.describe(include= ['O'])
```

Create simple relation plot

```
sns.relplot(data = df, x = 'timepoint', y = 'signal')
```

Create linear plot using relplot with confidential interval

```
sns.relplot(data = df, x = 'timepoint',y = 'signal',kind = 'line')
```

Create linear plot using relplot without confidential interval

```
sns.relplot(data = df, x = 'timepoint',y = 'signal',kind = 'line', ci = None)
```

Group data by table point and calculate mean of signal

```
grouped = df.groupby('timepoint').mean()
grouped
```

Plot grouped data

```
grouped.plot()
```

Show on plot point dependante of event

```
sns.relplot(data = df, x = 'timepoint', y = 'signal',hue = 'event')
```

Create plot with two lines dependante of event, with two columns.


```
sns.relplot(data = df, x = 'timepoint', y = 'signal',hue = 'event',
            kind = 'line',style = 'event',markers = True, col ='region')
```

Create plot with subjects column eneable columns wrap.

```
sns.relplot(data = df, x = 'timepoint', y = 'signal',hue = 'event',
            kind = 'line',style = 'event',markers = True, col ='subject',
            col_wrap = 2)
```
