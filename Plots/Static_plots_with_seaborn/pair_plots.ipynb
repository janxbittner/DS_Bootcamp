Importing modules.

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style = 'darkgrid',palette= 'viridis')
```

Import data about iris and show first 3 rows.

```
df = sns.load_dataset('iris')
df.head(3)
```

Create pari plot of whole data frame.

```
sns.pairplot(data = df)
```

Create the same graph as above but this time show spiecies in plots.

```
sns.pairplot(data = df, hue = 'species')
```

Create empyt pair plot.Then fill by kdeplots.

```
g = sns.PairGrid(df, hue = 'species')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot,n_levels = 9)
```
