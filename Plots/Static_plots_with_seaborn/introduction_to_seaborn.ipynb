Import modules

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

Generate 100 linear distributed values in range 0 -20

```
x = np.linspace(0,20,100)
y = np.random.randn(100,4)
y = np.cumsum(y,axis = 0)
```

Create linear plot with x,y values

```
plt.plot(x,y)
```

Set seaborn module to improve plots

```
sns.set()
```

Create again plot with x,y values

```
plt.plot(x,y)
```

Load data set from seaborn module, show first 5 rows

```
df = sns.load_dataset('tips')
df.head(n = 5)
```

Check if are any null values and if values are in logic range

```
df.describe()
```

Show information about categorical data in data frame

```
df.describe(include = ['category'])
```

```
sns.relplot(data = df, x = 'total_bill', y = 'tip')
```

Change default seaborn style

```
sns.set(style = 'whitegrid', palette= 'colorblind', font_scale= 1.1)
```

Create the same plot as above but include sex

```
sns.relplot(data = df, x = 'total_bill', y = 'tip', hue = 'sex')
```

Create plot which inlude sizes( number of pearson sittning in table)

```
sns.relplot(data = df, x = 'total_bill', y = 'tip', size= 'size', 
            sizes = (20,200), hue = 'size')
```

Create plot which show total_bill and tip dependent of size and time

```
sns.relplot(data = df,x = 'total_bill',y ='tip', col = 'time', 
            size = 'size',sizes = (20,200), hue = 'size',
            palette = 'viridis',row = 'smoker')
```
