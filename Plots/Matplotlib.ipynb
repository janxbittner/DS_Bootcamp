Import modules:

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Set seaborn to better plots:

```
sns.set()
```

Empty figure:

```
fig = plt.figure(1)
fig.show()
```

# Simply line plot:

```
plt.plot([0,1,2,3,4],
         [0,1,4,9,16],
                color='green',
                label = 'Power function')
plt.ylabel('Result')
plt.xlabel('Given value')
plt.title('Power of 2 function')
plt.legend()
_ = plt.show()
```

# Sinus and cosinus function:

```
plt.plot(np.arange(-3,3,0.1),
         np.sin(np.arange(-3,3,0.1)),
                color='blue',
                label = 'Sinus function')
plt.plot(np.arange(-3,3,0.1),
         np.cos(np.arange(-3,3,0.1)),
                color='green',
                label = 'Cosinus function')
plt.ylabel('Result')
plt.xlabel('Given value')
plt.title('Sinus (x) function')
plt.legend()
```

# Subplot Power functions:

```
plt.figure(figsize=(10,10))
plt.suptitle('Power functions')

plt.subplot(221)
plt.plot(np.arange(10),
         np.power(np.arange(10),2),
                label = 'Power of 2',
         color = 'red')
plt.title('Power of 2 Function')
plt.legend()

plt.subplot(222)
plt.plot(np.arange(10),
         np.power(np.arange(10),3),
                label = 'Power of 3',
         color ='blue')
plt.title('Power of 3 Function')
plt.legend()

plt.subplot(223)
plt.plot(np.arange(10),
         np.power(np.arange(10),4),
                label = 'Power of 4',
         color = 'brown')
plt.title('Power of 4 Function')
plt.legend()

plt.subplot(224)
plt.plot(np.arange(10),
         np.power(np.arange(10),5),
                label = 'Power of 5',
         color = 'green')
plt.title('Power of 5 Function')
plt.legend()
```
