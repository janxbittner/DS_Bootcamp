Importing modules.

```
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
```

Table about students and marks.

```
fig = go.Figure(data= [go.Table(header=dict(values = ['index','average mark'],
                                            fill_color = 'white',
                                            line_color = 'gray'),
                                cells = dict(values = [[287988,145228,304254,
                                                        250523],
                                         [3.8,3.9,3.2,4.1]],
                                         fill_color = 'lightblue'))],
                layout = go.Layout(width =500,height= 300))

fig.show() 
```
