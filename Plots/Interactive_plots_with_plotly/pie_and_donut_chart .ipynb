Importing modules.

```
import plotly.express as px
import pandas as pd
import seaborn as sns
import plotly.graph_objects as go
```

Load data about dimond using seaborn and show first 3 rows.

```
data = sns.load_dataset('diamonds')
data.head(3)
```

Show statistics about numerical data.

```
data.describe()
```

Show description about categorical data.

```
data.describe(include=['category'])
```

Dataframe with informations about types and quality of dimonds cut.

```
df_cuts = data['cut'].value_counts().reset_index()
df_cuts.columns = ['TYPE OF CUT', 'CUTS']
df_cuts
```

Circular plot with visualization of types and quality of diamonds cut.

```
import plotly.graph_objects as go

fig = go.Figure(data = [go.Pie(labels = df_cuts['TYPE OF CUT'],
                               values= df_cuts.CUTS)])
fig.show()
```

Ring plot with visualization of types and quality of diamonds cut.

```
go.Figure(data = [go.Pie(labels = df_cuts['TYPE OF CUT'],
                               values= df_cuts.CUTS,
                                 hole = 0.3,
                         title = 'Types of cuts'.upper(),
                         )])
```
