Importing modules.

```python
import plotly.express as px
import pandas as pd
import seaborn as sns
```

Load data about suvivers from titanic using seaborn and show first 3 rows.

```python
data = sns.load_dataset('titanic')
data.head(3)
```

Show statistics about numerical data.

```python
data.describe()
```

Show description about categorical data.

```python
data.describe(include = ['O'])
```

Scatter plot with age, sex and fare information.

```python
fig_sc = px.scatter(data,'age','fare', color = 'survived', facet_col= 'sex')
fig_sc.show()
```

Scatter plot with age, sex, class and fare information.

```python
fig_st2 = px.scatter(data,'age','fare', color = 'survived', facet_col= 'sex',
           facet_row = 'pclass', category_orders =
           {'pclass':[1,2,3],'sex':['female','male']})
fig_st2.show()
```
