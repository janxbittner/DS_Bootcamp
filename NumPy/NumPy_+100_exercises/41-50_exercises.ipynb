### <a name='3'></a> Exercise 43
Table _name_ from the previous exercise is shape of _(200, 300, 3)_. Make:
* Extend shape of _image_ to: _(1, 200, 300, 3)_.

````python
image = np.random.randint(low=0, high=256, size=(200, 300, 3), dtype=np.uint8)
print(image.shape)

# Solution
image = image.reshape(1,200,300,3)
print(image.shape)
````

````python
image = np.random.randint(low=0, high=256, size=(200, 300, 3), dtype=np.uint8)
print(image.shape)

# Solution
print(image.reshape(200,300,3,1).shape)
````

### <a name='4'></a> Exercise 44
There are two random generated tables _image1_ and _image2_. 


```
image1 = np.random.randint(low=0, high=256, size=(200, 300, 3), dtype=np.uint8)
image2 = np.random.randint(low=0, high=256, size=(200, 300, 3), dtype=np.uint8)
```

Each of this table extend and connect to variable _images_. Excepted size of result table: _(2, 200, 300, 3)_.

````python
image1 = np.random.randint(low=0, high=256, size=(200, 300, 3), dtype=np.uint8)
image2 = np.random.randint(low=0, high=256, size=(200, 300, 3), dtype=np.uint8)

# Solution
images =  np.concatenate((image1.reshape(1,200,300,3), image2.reshape(1,200,300,3)))
images.shape
````
