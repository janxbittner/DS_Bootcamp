Import module:

```
import numpy as np
```

# One-dimensional array:

Create a one-dimensional array:

```
one_dim = np.array([1,2])
```

Methods to show array:

```
one_dim
```

```
print(one_dim)
```

Type of one-dimensional array:

```
type(one_dim)
```

Numbers of row in array:

```
one_dim.ndim
```

Numbers of element in array:

```
one_dim.size
```

Dimension of array:

```
one_dim.shape
```

Data type in array:

```
one_dim.dtype
```
