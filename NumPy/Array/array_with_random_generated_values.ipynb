Import module:

```
import numpy as np
```

Set always the same random values:

```
np.random.seed(0)
```

Array with normal distributed 10 elements:

```
np.random.randn(10)
```

Array with normal distributed values of shape 2 rows and 5 columns:

```
np.random.randn(2,5)
```

Array with values between 0 - 1 of shape 2 array 2 rows and 5 columns:

```
np.random.rand(2,2,5)
```

Array with  intiger values between 11 - 22 of shape 2 rows and 5 columns:

```
np.random.randint(
    low =11,
    high = 23,
    size = (2,2,5)
)
```

Array with text values choosed randomly from list to create array shape of (2,3,4):

```
np.random.choice(['sql','excel','pandas'],size = (2,3,4))
```
