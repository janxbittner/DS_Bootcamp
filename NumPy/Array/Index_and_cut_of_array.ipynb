Import numpy module:

```
import numpy as np
```

Sample array:

```
A = np.arange(20)
A
```

# Indexing array:

Value of index number 5:

```
A[5]
```

Values of index number from 3 to 9:

```
A[3:9]
```

First 10 values from array:

```
A[:10]
```

Last 3 values from array:

```
A[-3:]
```

Ever third values from array starting from index number 2 and finish on index number 15:

```
A[2:15:3]
```

Array reshape to 4 rows and 5 columns:

```
A = A.reshape(4,5)
A
```

Cutting values from 6 to 8 and 11 to 13:

```
A[1:3,1:4]
```

Replace value 7 with -5:

```
A[1,2] = -5
```

Array after replace:

```
A
```

Reshape array to have 2 tables with 5 rows and 2 columns:

```
A = A.reshape(2,5,2)
A
```

The second table from array:

```
A[1]
```

Create more complicated array:

```
B = np.random.randint(low = -10, high = 10, size = (3,5,5))
B
```

Show the value from middle of second table:

```
B[1,2,2]
```

Show every second value in array:

```
B[::2,::2,::2]
```
