Importing module:

```
import numpy as np
```

## Operation on the same size array:

Create arrays:

```
A = np.array([2,5,-2,3])
B = np.array([3,-7,-2,0])
```

Addition:

```
A+B
```

```
np.add(A,B)
```

Subtraction:

```
A-B
```

```
np.subtract(A,B)
```

Multiplication:

```
A*B
```

```
np.multiply(A,B)
```

Division(Cause error because one value in zero):

```
A/B
```

```
np.divide(A,B)
```

```
(np.abs(A) * 3 + B) ** 2
```
