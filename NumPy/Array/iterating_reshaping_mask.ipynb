Import NumPy module:

```
import numpy as np
```

Create sample array:

# Iterating through array:

Print all tables in array with its number:


```
number = 0
for table in A:
    number +=1
    print(f"Table number: {number}")
    print(table,end='\n\n')
```

Print all third rows in array:

```
number = 0
for table in A:
    number +=1
    print(f"Table number: {number}, Row number: 3")
    print(table[2],end='\n\n')
```

Print all third columns in array:

```
number = 0
for table in A:
    number +=1
    print(f"Table number: {number}, Column number: 3")
    print(table[:,2],end='\n\n')
```

Print value at row number 4 and column 2 for each table:

```
number = 0
for table in A:
    number +=1
    print(f"Table number: {number}, Row number: 4, Column number: 2")
    print(f'Value is: {table[4,2]}',end='\n\n')
```

Print every values with are devided by 10:

```
number = 0
for item in A.flat:
    if item % 10 == 0:
        number += 1
        print(f'{number} element with value: {item}')
```

# Logic masks:

Array to test logic mask:

```
L = np.arange(start = -10, stop = 10, step = 0.5).reshape(10,-1)
L
```

Mask to check where values in array are lover then 0:

```
L < 0
```

Show only values lower then 0 in array using mask:

```
L[L<0]
```

Values between -5 and 5 by two methods:

```
L[(L>= - 5) & (L<= 5)]
```

```
L[np.bitwise_and(L >= -5,L<= 5)]
```
