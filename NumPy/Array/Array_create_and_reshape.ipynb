## Import modules:

```
import numpy as np
```

# Various method of create array:

# **Array filled by 0:**

```
zero_array = np.zeros(shape= (2,3,4),dtype = 'int8')
```

```
zero_array
```

# Array filled by 1:

```
ones_array = np.ones(shape = (2,3,4),dtype = 'uint8')
```

```
ones_array
```

# Array filed by 5:

```
fives_array = np.full(shape = (2,3,4), fill_value=5,dtype='float')
```

```
fives_array
```

```
np.linspace(0,4,5)
```

# Array with values from 0 to 8:

Using arrange:

```
np.arange(9)
```

```
np.arange(
    start = 0,
    stop = 9,
    step = 1,
    dtype = 'int'
)
```

Using linspace:

```
np.linspace(start = 0,
            stop = 8,
            num = 9,
            dtype = 'int')
```
