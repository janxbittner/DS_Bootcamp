# System of equations:

Example of equations:

2x	+	y	−	2z	=	3

x	−	y	−	z	=	0

x	+	y	+	3z	=	12

Create array of shells:

```
A = np.array([[2,1,-2],
              [1,-1,-1],
              [1,1,3]])
A
```

Check if numbers of column and rows is the same:

```
print('Yes') if A.shape[0] ==A.shape[1] else print("No")
```

Row and column names are the same so we can calculate the inverse of a matrix A:

```
A_inv = np.linalg.inv(A)
A_inv
```

Create array of results:

```
B = np.array([[3],[0],[12]])
B
```

Compute the solution of equations:

```
A_inv @ B
```

Solution:

```
letter = ['x','y','z']
for idx, val in enumerate(A_inv @ B):
    print(f'{letter[idx]} is equal: {val}')
```
