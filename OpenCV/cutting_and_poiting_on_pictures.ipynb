Importing libraries:

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

Load and show images:

```python
from PIL import Image
import requests

view_url = 'https://i.pinimg.com/originals/c6/13/\
06/c61306712e50726c2ae3a093f29105a7.jpg'
view = np.asarray(Image.open(requests.get(view_url, stream=True).raw))
imgplot = plt.imshow(view)
plt.title('First picture')
plt.axis('off')
_= plt.show()
```

```python
from PIL import Image
import requests

bear_url = 'https://video.cgtn.com/news/7a596a4e784d544e\
33516a4d77457a4e786b444f31457a6333566d54/video/f4f4d4a40e\
0743b1a1e347ab0dbe6085/f4f4d4a40e0743b1a1e347ab0dbe6085.jpg'

bear_img = np.asarray(Image.open(requests.get(bear_url, stream=True).raw))
imgplot = plt.imshow(bear_img)
plt.title('Second picture')
plt.axis('off')
_= plt.show()
```

Size of pictures:

```python
print(f'Bear image is shape of: {bear_img.shape}')
print(f'View image is shape of: {view.shape}')
```

Convert picture to gray scale:

```python
viev_gray = cv2.cvtColor(view, cv2.COLOR_BGR2GRAY)

imgplot = plt.imshow(viev_gray,cmap = 'gray')
plt.title('View in gray scale')
plt.axis('off')
_= plt.show()
```

Cutting the edges of picture:

```python
left_top = viev_gray[:200, :200]

imgplot = plt.imshow(left_top,cmap = 'gray')
plt.title('View of left top corner in gray scale')
plt.axis('off')
_= plt.show()
```

```python
right_top = viev_gray[:200, -200:]

imgplot = plt.imshow(right_top,cmap = 'gray')
plt.title('View of right top corner in gray scale')
plt.axis('off')
_= plt.show()
```

```python
left_bottom = viev_gray[-200:, :200]

imgplot = plt.imshow(left_bottom,cmap = 'gray')
plt.title('View of left bottom corner in gray scale')
plt.axis('off')
_= plt.show()
```

```python
right_bottom = viev_gray[-200:, -200:]

imgplot = plt.imshow(right_bottom,cmap = 'gray')
plt.title('View of right bottom corner in gray scale')
plt.axis('off')
_= plt.show()
```

```python
hight,width = viev_gray.shape

center_square = viev_gray[int(hight/2-100):int(hight/2+100),
                          int(width/2-100):int(width/2+100)]

imgplot = plt.imshow(center_square,cmap = 'gray')
plt.title('View of center square in gray scale')
plt.axis('off')
_= plt.show()
```

Resize pictures:

```python
import imutils
small_center_square = imutils.resize(image = center_square, height = 20)

imgplot = plt.imshow(small_center_square,cmap = 'gray')
plt.title('View of resized center square in gray scale')
plt.axis('off')
_= plt.show()
```

Show edges of bear picture:

```python
imgplot = plt.imshow(cv2.Laplacian(src = bear_img,ddepth = cv2.CV_64F).astype('uint8'))
plt.title('Viev of edges of bear pictures usingLaplacian')
plt.axis('off')
_= plt.show()
```

```python
imgplot = plt.imshow(cv2.Canny(bear_img,200,200))
plt.title('Viev of edges of bear pictures using Canny')
plt.axis('off')
_= plt.show()
```

Painting on picture:

```python
start_point = (250,250)
end_point = (1000,1000)
color = (0, 0, 255)
thickness = 50


imgplot = plt.imshow(cv2.rectangle(bear_img.copy(), start_point, end_point, color, thickness)) 
plt.title('Viev of bear pictures with rectangle')
plt.axis('off')
_= plt.show()
```

```python
start_point = (4300,1800)
radius = 500
color = (0, 0, 255)
thickness = 50


imgplot = plt.imshow(cv2.circle(bear_img.copy(), start_point, radius, color, thickness)) 
plt.title('Viev of bear pictures with rectangle')
plt.axis('off')
_= plt.show()
```

```python
start_point = (4300,1800)
radius = 500
color = (0, 0, 255)
thickness = -1


imgplot = plt.imshow(cv2.circle(bear_img.copy(), start_point, radius, color, thickness)) 
plt.title('Viev of bear pictures with rectangle')
plt.axis('off')
_= plt.show()
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX 
org = (200, 700)  
fontScale = 25
color = (0, 0, 255) 
thickness = 8

imgplot = plt.imshow(cv2.putText(bear_img.copy(), 'Bear', org, font,  
                   fontScale, color, thickness, cv2.LINE_AA)) 
plt.title('Viev of bear pictures with text')
plt.axis('off')
_= plt.show()
```
