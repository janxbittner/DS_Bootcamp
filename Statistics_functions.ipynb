Import modules:

```
import pandas as pd
import pandas_datareader as pdr
import numpy as np
```

# Get data:

Import Amazon stock data: 

```
df = pdr.DataReader(name='AMZN', data_source='yahoo')
df
```

Remove last column:

```
df = df.iloc[:,:-1] 
df
```

Take only last 10 rows:

```
df = df.iloc[-10:,:] 
df
```

Create array with df values:

```
A = df.to_numpy()
A
```

## Statistics_functions

Sum of volume:

```
A[:,-1].sum()
```

Calulate sum of each column:

```
A.sum(axis = 0)
```

Calulate sum of each row:

```
A.sum(axis = 1)
```

Print minimum and maximum value for each column:

```
for num in range(0,A.shape[1]):
    print(f'For column number: {num}, Max value is: {A[:,num].max()}\
    , Min value is: {A[:,num].min()}')
```

Compute mean, median, std and var for each column:

```
for num in range(0,A.shape[1]):
    print('For column: ' + str(num))
    print(f'Mean: {np.mean(A,axis = 0)[num]}, Median: {np.median(A,0)[num]} '\
          + f'Std: {np.std(A,0)[num]}, Var: {np.var(A,0)[num]}', end = '\n\n')
```
