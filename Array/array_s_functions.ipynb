Importing module:

```
import numpy as np
```

Create random array:

```
A = np.random.randn(2,3,5)*10
A
```

# Mathematic functions:

Expotencial of array:

```
np.exp(A)
```

Square root of array:

```
np.sqrt(abs(A))
```

Maximal value in array:

```
np.max(A)
```

Minimal value in array:

```
np.min(A)
```

Mean of array:

```
np.mean(A)
```

Median of array:

```
np.median(A)
```

Standard deviation of array:

```
np.std(A)
```

# Logical functions:

Check if is all values are diffrent then 0:

```
np.all(A)
```

```
np.all([1,2,0,3])
```

Check if all values are not 0:

```
np.any(A)
```

```
np.any([0,0])
```

# Index functions:

Examped array:

```
np.arange(5)
```

Index number of lowest value in array:

```
np.argmin(np.arange(5))
```

The lowest value:

```
np.arange(5)[np.argmin(np.arange(5))]
```

Index number of highest value in array:

```
np.argmax(np.arange(5))
```

The highest value:

```
np.arange(5)[np.argmax(np.arange(5))]
```

Array with sorted index number from lowest to highest value:


```
np.argsort(np.arange(5))
```

Array with sorted index number from highest to lowest value:


```
np.argsort(np.arange(5))[::-1]
```
