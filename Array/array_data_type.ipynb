Import module:

```
import numpy as np
```

# Array with intiger values:

```
array_int = np.array([[1,2],[2,8]])
```

Methods to show array:

```
array_int
```

```
print(array_int)
```

Type of values in array:

```
array_int.dtype
```

#Array with float values:

```
array_float = np.array([[1.3,2.5],[2.2,8.8]])
```

Methods to show array:

```
array_float
```

```
print(array_float)
```

Type of values in array:

```
array_float.dtype
```

#Array with int values converted to float:

```
array_float_conv = np.array([[1,3,2,5],[2,2,8,8]],dtype='float')
```

Methods to show array:

```
array_float_conv
```

```
print(array_float_conv)
```

Type of values in array:

```
array_float_conv.dtype
```
