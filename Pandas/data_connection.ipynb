Import modules:

```
import pandas as pd
import numpy as np
import pandas_datareader.data as web
from datetime import date
```

# Import data:

Load stocks data to seperated dataframe:

```
start = '01-01-2010'
end = date.today()
Uber = web.DataReader("UBER", 'yahoo', start, end)
Google = web.DataReader("GOOGL", 'yahoo', start, end)
Amazon = web.DataReader("AMZN", 'yahoo', start, end)
Apple = web.DataReader("AAPL", 'yahoo', start, end)
```

# Data processing:

Make unique columns name:

```
Uber.columns = [('uber_' + col.lower()).replace(' ','_') for col in Uber.columns]
Uber.columns
```

```
Amazon.columns = [('amazon_' + col.lower()).replace(' ','_') for col in Amazon.columns]
Amazon.columns
```

```
Apple.columns = [('apple_' + col.lower()).replace(' ','_') for col in Apple.columns]
Apple.columns
```

```
Google.columns = [('google_' + col.lower()).replace(' ','_') for col in Google.columns]
Google.columns
```

# Data connection:

Concat Method:

```
con_df = pd.concat(objs=[Amazon,Apple,Google,Uber],
          axis = 1,
          join = 'outer')

con_df
```

Statistics about joined dataframe:

```
pd.set_option('display.float_format',lambda x: f'{x:.2f}')
con_df.describe().T
```

Calculate correlation matrix for all close columns:

```
con_df_close = con_df[con_df.columns[con_df.columns.str.endswith('close')]]
col_list = con_df_close[con_df.columns[con_df.columns.str.endswith("adj_close")]].columns
con_df_close = con_df_close.drop(columns=col_list)
con_df_close.corr()
```
