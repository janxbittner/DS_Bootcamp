Import modules:

```
import pandas as pd
import numpy as np
```

Sample series:

```
s = pd.Series(
    data = np.random.randint(-100,100,100),
    index =np.arange(100),
    dtype = 'float32',
    name = 'Random values'
)

s
```

Change random value from series into numPy nan:

```
s[np.random.randint(low = 1,high = 100,size = 1)] = np.NAN
```

Show NAN value in series:

```
print(s[s.isna()])
print("Where number stands for index number and NaN is not definded value")
```

Series where index is date and values are numbers of day in year:

```
cal = pd.Series(data= np.arange(1,367,1,dtype= 'int16'),
                index = pd.date_range(
                    start = '01-01-2020',
                    end = '31-12-2020',
                    freq = 'D'
                ), name = 'Days of year')

cal
```

Series from dictionary:

```
dictionary = dict(Adam = 21, Lisa = 43, Jack = 33, Lena = 19)
dictionary
```

```
s_dict = pd.Series(data = dictionary,
                   name = 'Name and age',
                   dtype = 'int8')

s_dict
```

Return age of person:

```
p_name = input('Enter name:\n')
if p_name in s_dict.keys():
    print(f"{p_name} is {s_dict[p_name]} years old!")
```

```
s = s.sort_values()
s
```

```
s = s.apply(lambda x: np.sqrt(np.abs(x)))
s
```
