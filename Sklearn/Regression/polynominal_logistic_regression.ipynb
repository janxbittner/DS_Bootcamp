# Polynominal logistic regression:

Import modules:

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Create data:

```
X = np.arange(-10,10,0.5)
noise = 80 * np.random.randn(40)
y = -X**3 + 10*X**2 - 2*X + 3 + noise
X= X.reshape(40,1)
_ = plt.scatter(X,y)
```

Make model of linear regression:

```
from sklearn.linear_model import LinearRegression
model_1 = LinearRegression()
model_1.fit(X,y)
y_predict = model_1.predict(X)

plt.scatter(X,y,label = 'Real values')
plt.plot(X,y_predict, label = 'Regression 1 dimens.', c = 'red')
_ = plt.legend()
```

```
from sklearn.metrics import r2_score
print(f'R2 Score is equal:{r2_score(y,y_predict):.2f} %')
```

Polynominal regression:

```
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X)
X_poly[:10]
```

```
model_poly_2 = LinearRegression()
model_poly_2.fit(X_poly,y)
y_poly_2 = model_poly_2.predict(X_poly)

```

```
plt.scatter(X,y,label = 'Real values')
plt.plot(X,y_predict, label = 'Regression 1 dimens.', c = 'red')
plt.plot(X,y_poly_2, label = 'Polinominal Regression 2 deg.', c = 'blue')
_ = plt.legend()
```

```
print(f'R2 Score is equal:{r2_score(y,y_poly_2):.2f} %')
```

For 3 degree:

```
poly_3 = PolynomialFeatures(degree=3)
X_poly_3 = poly_3.fit_transform(X)

model_poly_3 = LinearRegression()
model_poly_3.fit(X_poly_3,y)
Y_result_3 = model_poly_3.predict(X_poly_3)

plt.scatter(X,y,label = 'Real values')
plt.plot(X,y_predict, label = 'Regression 1 dimens.', c = 'red')
plt.plot(X,y_poly_2, label = 'Polinominal Regression 2 deg.', c = 'blue')
plt.plot(X,Y_result_3, label = 'Polinominal Regression 3 deg.', c = 'violet')
_ = plt.legend()
```

```
print(f'R2 Score is equal:{r2_score(y,Y_result_3):.2f} %')
```
